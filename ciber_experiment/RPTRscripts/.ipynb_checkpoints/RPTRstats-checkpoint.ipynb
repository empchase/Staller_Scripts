```python
import string
s = 'string] with. punctuation'
string.punctuation
```

```python
filename = 'slurm-18652952.out'

libstats = {}

with open(filename, 'r') as f:
    
    for line in f:
        liblist = []
        if line.startswith('['):
            elements = line.split(",")
            key = '_'.join(elements[0].strip(string.punctuation).split('_')[0:3])
            totreads = elements[1]
            goodlen = elements [2]
            matches = elements[-1].strip().strip(string.punctuation)
            unique = elements[-2]
            liblist.extend([key, totreads, goodlen, matches, unique])
            libstats[key] = liblist
            
print(libstats)
    
```

```python
rptrstats = pd.DataFrame.from_dict(libstats)
rptrstats.head(20)
```
